# Scholarship Eligibility Checker

# Dictionary of names and ages

```python
# Names and ages of candidates 

candidates = {
    "Sam:": 17,
    "Cate": 19,
    "Nthabi": 22,
    "Winnie": 18,
    "Frank": 20,
    "Tracey": 21,
}
```

```python
# Initialize a list to store eligible candidates

eligible = []
# Loop through the dictionary using .items()
for name, age in candidates.items():
    
    # Condition to check eligibility
    if age > 18:
        eligible.append(name)

# Print results
print("Scholarship Eligible Candidates:")
for name in eligible:
    print(f"- {name}")
```

# Bar chat to illustrate

```python
import matplotlib.pyplot as plt
plt.bar(candidates.keys(), candidates.values(), color=['green' if age > 18 else 'red' for age in candidates.values()])
plt.axhline(y=18, color='blue', linestyle='--', label="Eligibility Threshold")
plt.title("Scholarship Eligibility Check")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
